```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

# Define el modelo en la forma de un sistema de ODEs
def modelo_de_competencia(z, t):
    x, y = z
    dxdt = x * (1 - 0.1*x - 0.05*y)
    dydt = y * (1.7 - 0.1*y - 0.15*x)
    return [dxdt, dydt]

# Vector de tiempo para 0 a 200 aos
t = np.linspace(0, 200, 1000)

# Condiciones iniciales para el caso a
inicial_condiciones_a = [1.1, 1]

# Resolver ODE para el caso a
solucion_a = odeint(modelo_de_competencia, inicial_condiciones_a, t)

# Condiciones iniciales para el caso b
inicial_condiciones_b = [4, 10]

# Resolver ODE para el caso b
solucion_b = odeint(modelo_de_competencia, inicial_condiciones_b, t)

# Graficar resultados para el caso a
plt.figure(figsize=(10, 5))
plt.plot(t, solucion_a[:, 0], label='x(t)')
plt.plot(t, solucion_a[:, 1], label='y(t)')
plt.title('Caso a: x(0) = 1.1, y(0) = 1')
plt.xlabel('Tiempo (aos)')
plt.ylabel('Poblaciones (miles)')
plt.legend()
plt.grid(True)
plt.show()

# Graficar resultados para el caso b
plt.figure(figsize=(10, 5))
plt.plot(t, solucion_b[:, 0], label='x(t)')
plt.plot(t, solucion_b[:, 1], label='y(t)')
plt.title('Caso b: x(0) = 4, y(0) = 10')
plt.xlabel('Tiempo (aos)')
plt.ylabel('Poblaciones (miles)')
plt.legend()
plt.grid(True)
plt.show()
```
